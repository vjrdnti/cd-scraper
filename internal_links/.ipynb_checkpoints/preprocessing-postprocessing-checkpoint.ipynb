```python
i=0
def check(row):
    try:
        if len(row.split('web-api/')[1].split('/'))==3:
            a = 'https://'
            li = row.split('/')
            for i in range(2,len(li)-1):
                if i==len(li)-2:
                    a+=str(li[i])
                else:
                    a += str(li[i]) + '/'
            return a
        else:
            return row
    except:
        print('correct row:')
        print('***************')
        return a
df['link'] = df['link'].apply(check)
```

```python
df = pd.read_csv('./main.csv')
```

```python
coll = list(df.columns)
```

```python
import ast

dfn = []
for idx,row in df.iterrows():
    ans = {}
    for c in coll:
        if c=='name' or c=='institute type':
            ans[f'{c}'] = row[f'{c}']
        try:
            ans[f'{c}'] = ast.literal_eval(row[f'{c}'])
        except:
            ans[f'{c}'] = row[f'{c}']
    dfn.append(ans)
df = pd.DataFrame(dfn)
```

```python
new_df = []
for idx, row in df.iterrows():
    new = {}
    
    try:
        new['name'] = row['name']
    except:
        new['name'] = pd.NA

    try:
        new['institute type'] = row['institute type']
    except:
        new['institute type'] = pd.NA

    try:
        new['id'] = row['basic_info']['id']
    except:
        new['id'] = pd.NA

    try:
        new['year_founded'] = row['basic_info']['year_founded']
    except:
        new['year_founded'] = pd.NA

    try:
        new['city'] = row['basic_info']['city']
    except:
        new['city'] = pd.NA

    try:
        new['state'] = row['basic_info']['state']
    except:
        new['state'] = pd.NA

    try:
        new['university_type'] = row['basic_info']['university_type']
    except:
        new['university_type'] = pd.NA

    try:
        new['address'] = row['basic_info']['address']['address']
    except:
        new['address'] = pd.NA

    try:
        new['pincode'] = row['basic_info']['address']['pincode']
    except:
        new['pincode'] = pd.NA

    try:
        new['mobile1'] = row['basic_info']['mobile']
    except:
        new['mobile1'] = pd.NA

    try:
        new['mobile2'] = row['basic_info']['phone_no']
    except:
        new['mobile2'] = pd.NA

    try:
        new['website'] = row['basic_info']['website']
    except:
        new['website'] = pd.NA

    try:
        new['approved_by'] = row['basic_info']['approved_by']
    except:
        new['approved_by'] = pd.NA

    try:
        new['major_stream'] = row['basic_info']['major_stream']
    except:
        new['major_stream'] = pd.NA

    try:
        new['major_stream_name'] = row['basic_info']['major_stream_name']
    except:
        new['major_stream_name'] = pd.NA

    try:
        new['affiliated_to'] = row['basic_info']['affiliated_to']['name']
    except:
        new['affiliated_to'] = pd.NA
        print(idx)

    facilities = []
    try:
        for i in row['facilities']:
            facilities.append(i['name'])
        new['facilities'] = facilities
        new['facility_count'] = len(facilities)
    except:
        new['facilities'] = pd.NA
        new['facility_count'] = pd.NA

    try:
        for i in row['course']:
            new['stream'] = i['stream_data']['name']
            new['stream_id'] = i['stream_data']['id']
            new['course'] = i['display_name']
            new['duration'] = i['duration_year']
            new['level'] = i['level']
            new['course_type'] = i['course_type']
            new['mode'] = i['degree_could_be']
            new['eligibility'] = i['eligibility']
            new['admission_criteria'] = i['admission_criteria']
            for k in i['streams']:
                new['substream'] = k['name']
                try:
                    new['fees'] = k['fees_data']['amount']
                except:
                    new['fees'] = pd.NA

                try:
                    new['fee_additional'] = k['fees_data']['text']
                except:
                    new['fee_additional'] = pd.NA

                try:
                    new['cutoff'] = k['cutoff']['cutoff']
                except:
                    new['cutoff'] = pd.NA

                try:
                    new['exam'] = k['cutoff']['exam']
                except:
                    new['exam'] = pd.NA

                try:
                    new['cutoff_data_from'] = k['cutoff']['year']
                except:
                    new['cutoff_data_from'] = pd.NA
                
                new_df.append(new)
    except:
        new['stream'] = pd.NA
        new['stream_id'] = pd.NA
        new['course'] = pd.NA
        new['duration'] = pd.NA
        new['level'] = pd.NA
        new['course_type'] = pd.NA
        new['mode'] = pd.NA
        new['eligibility'] = pd.NA
        new['admission_criteria'] = pd.NA
        new['substream'] = pd.NA
        new['fees'] = pd.NA
        new['fee_additional'] = pd.NA
        new['cutoff'] = pd.NA
        new['exam'] = pd.NA
        new['cutoff_data_from'] = pd.NA
        new_df.append(new)
```

```python
df2 = pd.DataFrame(new_df)
```
